Question 4: How many columns have missing values?

Question 6. Median value of horsepower

```python
# Find the median value of the horsepower column in the dataset.
df['horsepower'].median()
```

```python
# Calculate the most frequent value of the same horsepower column.
df['horsepower'].mode()
```

```python
# Use fillna method to fill the missing values in the horsepower column with the most frequent value from the previous step
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].mode().to_list()[0])
```

```python
# Calculate the median value of horsepower again
df['horsepower'].median()
```

Question 7: Sum of Weights

```python
df_asia = df[df['origin'] == 'Asia'][['vehicle_weight', 'model_year']]
```

```python
df_asia.head(7)
```

```python
X = df_asia.head(7).to_numpy()
X
```

```python
XTX = np.dot(X.T, X)
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv.shape
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
```

```python
X.T.shape
```

```python
_w = np.dot(XTX_inv, X.T)
_w.shape
```

```python
w = np.dot(_w, y)
w.shape
```
